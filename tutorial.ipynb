```python
# manual machine learning The Hitchhiker's guide to phyton
from scipy.stats import ttest_ind
setosa = iris[iris['species']=='setosa']
virginica = iris[iris['species']=='virginica']

print(ttest_ind(setosa['sepal_length'], virginica['sepal_length']))
print(ttest_ind(setosa['sepal_length'], virginica['sepal_length']))
```

```python
# first cargar imagen anatomica disponible por defecto en el modulo Nilearn
%matplotlib inline
from nilearn import datasets
# defaul MNI152 template
anatomical_image = datasets.load_mni152_template()
# anatomica_image = "user/"

#luego se importa el modulo plotting de nilearn
from nilearn import plotting

plotting.plot_anat(anatomical_image, title="MNI TEMPLATE")

```

```python
# Cartgar estudio de RM_anatomia
localizer_dataset = datasets.fetch_localizer_button_task(
n_subjects=1,
get_anats= True)

localizer_anat_filename = localizer_dataset.anats[0]
print(localizer_anat_filename)
#T_contrast map filename

localizer_tmap_filename = localizer_dataset.tmaps[0]
print(localizer_tmap_filename)
```

```python
#visualizacion
plotting.plot_stat_map(localizer_tmap_filename, bg_img=localizer_anat_filename,
                      threshold=3, title="statical map with threshold")
```

```python
plotting.plot_glass_brain(localizer_tmap_filename, threshold=3,
                         plot_abs=False, colorbar=True,
                         title='Statical map plot_abs=False & threshols')
```

```python
#showing percentile the¿reshold image
plotting.plot_stat_map(localizer_tmap_filename,display_mode='z', threshold=3,
                      cut_coords=5, colorbar=True)
```
